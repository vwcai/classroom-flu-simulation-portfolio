# Classroom Flu Monte Carlo Simulation
This notebook compares two classroom outbreak scenarios over 5,000 Monte Carlo runs: one with no initial immunity and one with 50% initial immunity.

## Simulation 1: No Initial Immunity

```python
import numpy as np
import scipy
import pandas as pd
import matplotlib.pyplot as plt
from utils import (
    PandemicFluSimulation, 
    plot_separate_and_mean_combined_results, 
    hist_epidemic_durations
)
```

```python
NUM_RUNS = 5000
```

```python
  # number of simulation runs
SIMULATION_PARAMS_1 = {
    'num_kids': 61,
    'days_to_run': 63,
    'infection_prob': 0.01,
    'recovery_duration': 3,
    'immune_ratio': 0,
    'seed': 42
}
```

```python
sim_1 = PandemicFluSimulation(**SIMULATION_PARAMS_1)

# run monte carlo simulations
monte_carlo_results_1 = sim_1.run_monte_carlo(num_runs=NUM_RUNS)
```

## Simulation 1 Results

```python
plot_separate_and_mean_combined_results(monte_carlo_results_1)
```

```python
# Sim 1: simulation results of time to peak and epidemic duration 
monte_carlo_results_1['time_to_peak'], monte_carlo_results_1['epidemic_duration']
```

## Expected Accumulated Infections in Simulation 1

```python
accumulate_I_sim_1 = pd.DataFrame({
    'By Day': range(1, 64),
    'Accumulated Infections': monte_carlo_results_1['DeltaI']['mean'].cumsum() + 1,
})
accumulate_I_sim_1.T.to_csv('sim_1_expected_num_of_accumulated_infections.csv')
```

```python
display(accumulate_I_sim_1.T)
```

```python
plt.figure(figsize=(6, 2)) 
plt.plot(
    accumulate_I_sim_1['By Day'], 
    accumulate_I_sim_1['Accumulated Infections'], 
    marker='o', 
    markersize=3
)
plt.xlabel('Day', fontsize=10) 
plt.ylabel('number of kids', fontsize=10)
plt.title('Expected Number of Accumulated Infections, Initial Immune Rate=0', fontsize=10) 
plt.show()
```

## Epidemic Duration in Simulation 1

```python
hist_epidemic_durations(monte_carlo_results_1['all_raw_epidemic_durations'], bins=30)
```

## Early Infection Counts in Simulation 1

```python
print('expected number of infections by Day 1 including Tommy is',\
      1 + monte_carlo_results_1['DeltaI']['mean'][0])
print('expected number of infections by Day 2 including Tommy is',\
      1 + monte_carlo_results_1['DeltaI']['mean'][0] + monte_carlo_results_1['DeltaI']['mean'][1])
```

## Compare Infection Trajectories Across Scenarios

```python
plt.figure(figsize=(6, 2)) 


plt.plot(
    accumulate_I_sim_1['By Day'], 
    accumulate_I_sim_1['Accumulated Infections'], 
    marker='o', 
    markersize=3,
    label='Initial Immune Rate = 0' 
)


plt.plot(
    accumulate_I_sim_2['By Day'], 
    accumulate_I_sim_2['Accumulated Infections'], 
    marker='x', 
    markersize=3,
    color='red', 
    linestyle='--', 
    label='Initial Immune Rate = 0.5'
)

plt.xlabel('Day', fontsize=10) 
plt.ylabel('number of kids', fontsize=10)
plt.title('Expected Number of Accumulated Infections Comparison', fontsize=10) 

plt.legend(fontsize=7) 

plt.show()
```
